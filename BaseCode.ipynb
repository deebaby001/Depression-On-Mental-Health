```python
import pandas as pd
# Sample DataFrame with more than 10 records 
data = { 'Work Pressure': [5.0, 4.0,0.0,5.0,1.0], 'Study Satisfaction': [0.0, 0.0, 2.0, 0.0,0.0] }

df = pd.DataFrame(data)

# Select the top 5 records
top_5_records = df.head(5)

# Display the top 1000 records
print(top_5_records)
```

```python
import matplotlib.pyplot as plt

# Create a bar plot for the top 1000 records
plt.figure(figsize=(10, 6))
plt.bar(top_5_records['Work Pressure'], top_5_records['Study Satisfaction'], color='skyblue')
plt.xlabel('Work Pressure')
plt.ylabel('Study Satisfaction')
plt.title('Study Satisfaction by Work Pressure (Top 5 Records)')
plt.show()
```

From the Bar Graph,the cursory result is indicating that the higher the 'Study Satisfaction' indices, the lower the 'Work Pressure' indices.

```python
# Line Plot
import matplotlib.pyplot as plt 
x = [5.0, 4.0, 0.0, 5.0, 1.0] 
y = [0.0, 0.0, 2.0, 0.0, 0.0] 

plt.plot(x, y) 
plt.xlabel('Work Pressure') 
plt.ylabel('Study Satisfaction') 
plt.title('Line Plot Graph of Work Pressure and Study Satisfaction') 
plt.show()
```
